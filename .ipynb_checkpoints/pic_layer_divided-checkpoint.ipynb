### 图像特征提取实验

```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
# 读取图片路径
os.chdir(r"I:\__data__\XIMEA_2020-11-13\5_0.25\layer_9")
pic_list = os.listdir()
print(pic_list)
```

```python
# 判断图像是否符合要求
def pic_process(pic):
    img = cv2.imread(pic,cv2.IMREAD_GRAYSCALE)
    count = 0
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            if img[i,j] <= 244:                                          #要求
                count = count + 1
    if count >= img.shape[0]*img.shape[1]-100:
        return True
    else:
        return False
```

```python
img = cv2.imread(pic_list[0],cv2.COLOR_BGR2GRAY)
print(img.shape)
# b,g,r = cv2.split(img)
# print(b ,g ,r )
ret, th1 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_TRIANGLE )
print(ret)
cv2.imshow("roi image", th1)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.hist(img.ravel(),256)
plt.show()
```

```python
def maxnumpixel(pic):
    img = cv2.imread(pic)
    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    grayDict = {}
    for key in range(256):
        grayDict[key] = 0
    for i in range(img_gray.shape[0]):
        for j in range(img_gray.shape[1]):
            grayDict[img_gray[i][j]] += 1
    maxnumpixel = max(grayDict.values())
    print(maxnumpixel)
    numberOfmaxnum_pixel = filter(lambda x:maxnumpixel == x[1], grayDict.items())
    print(list(numberOfmaxnum_pixel)[0][0])
    plt.figure(dpi=300)
    plt.xticks(range(0,255,25))
    plt.bar(list(grayDict.keys()), list(grayDict.values()))
    plt.savefig(pic+"_hist.png")
    return list(numberOfmaxnum_pixel)[0][0]
```

```python
temp_list = []
for pic in pic_list:
    keynum = maxnumpixel(pic)
    temp_list.append(keynum)
print(temp_list)
```

```python
count = 0
for pic in pic_list:
    flag = pic_process(pic)
    if flag:
        count += 1
    print(count)
print(count)
```
